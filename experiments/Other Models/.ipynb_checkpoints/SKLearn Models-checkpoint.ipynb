# SKLearn Model Analysis

The purpose of this notebook is to train (and optimize) other models to see their accuracy with the idea of potentially using them, or their results to better our neural network

## Required Files

This notebook relies on the `breastmnist.npz` and `pneumoniamnist.npz` files. Their paths are in the *Load Data* section.

(Optional) The *Build Models* section trains a variety of models and stores them in a `dict` called `models`. However, at the very end the models are exported into a file called `sklearn_models.joblib` that can be loaded with `joblib.load` an 

```python
from collections import defaultdict
from tqdm import tqdm
import itertools
from joblib import Parallel, delayed, dump, load

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
```

## Utilities

This function helps process the image dataset into something usable by the `fit` and `predict` models employed by `sklearn`

```python
def process_image_arr_sklearn(images):
    return images.reshape(len(images), -1)
```

This function computes the ROC curve of models given some data

```python
def roc(model, X, y):
    y_score = model.predict_proba(X)[:, 1]
    return roc_auc_score(y, y_score)
```

## Load Data

```python
breast = np.load("../../data/breastmnist.npz")

breast_trainX = process_image_arr_sklearn(breast["train_images"])
breast_trainY = breast["train_labels"].flatten()

breast_valX = process_image_arr_sklearn(breast["val_images"])
breast_valY = breast["val_labels"].flatten()

breast_testX = process_image_arr_sklearn(breast["test_images"])
breast_testY = breast["test_labels"].flatten()
```

```python
pneumonia = np.load("../../data/pneumoniamnist.npz")

pneumonia_trainX = process_image_arr_sklearn(pneumonia["train_images"])
pneumonia_trainY = pneumonia["train_labels"].flatten()

pneumonia_valX = process_image_arr_sklearn(pneumonia["val_images"])
pneumonia_valY = pneumonia["val_labels"].flatten()

pneumonia_testX = process_image_arr_sklearn(pneumonia["test_images"])
pneumonia_testY = pneumonia["test_labels"].flatten()
```

## Build Models

- Naive Bayes
- Decision Tree
- KNN
- Logistic Regression
- SVM


```python
models = defaultdict(lambda: {"breast": None, "pneumonia": None})
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

#### Model Accuracy

```python
models["naive_bayes"]["breast"] = GaussianNB()
models["naive_bayes"]["breast"].fit(breast_trainX, breast_trainY)

models["naive_bayes"]["pneumonia"] = GaussianNB()
models["naive_bayes"]["pneumonia"].fit(pneumonia_trainX, pneumonia_trainY)
```

```python
print(f'breast train acc: {models["naive_bayes"]["breast"].score(breast_trainX, breast_trainY)}')
print(f'breast test acc: {models["naive_bayes"]["breast"].score(breast_testX, breast_testY)}')
print(f'breast val acc: {models["naive_bayes"]["breast"].score(breast_valX, breast_valY)}')
print()
print(f'pneumonia train acc: {models["naive_bayes"]["pneumonia"].score(pneumonia_trainX, pneumonia_trainY)}')
print(f'pneumonia test acc: {models["naive_bayes"]["pneumonia"].score(pneumonia_testX, pneumonia_testY)}')
print(f'pneumonia val acc: {models["naive_bayes"]["breast"].score(pneumonia_valX, pneumonia_valY)}')
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Hyperparameters to Optimize

- `criterion` {"gini", "entropy", "log_loss"}
- `min_samples_split` - Number of samples that warrant a leaf splitting (indirectly controls `max_depth`)
- `min_samples_leaf` - Number of samples required to be a leaf node?

```python
def eval_params(criterion, min_samples_split, min_samples_leaf):
    model = DecisionTreeClassifier(criterion="log_loss", min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
    return cross_val_score(model, breast_trainX, breast_trainY, scoring="roc_auc", cv=5, n_jobs=-1).mean()

params = itertools.product(
    ("gini", "entropy", "log_loss"),
    range(2, 50),
    range(1, 20)
)

results = Parallel(n_jobs=-1, verbose=10)(delayed(eval_params)(criterion, min_samples_split, min_samples_leaf) for criterion, min_samples_split, min_samples_leaf in params)
```

```python
params = list(itertools.product(
    ("gini", "entropy", "log_loss"),
    range(2, 50),
    range(1, 20)
))

best_score_index = results.index(max(results))
best_score = results[best_score_index]
best_params = params[best_score_index]

print(f"Params: {best_params}")
print(f"Score: {best_score}")
```

#### Model Accuracy

```python
models["decision_tree"]["breast"] = DecisionTreeClassifier(criterion="gini", min_samples_split=3, min_samples_leaf=10)
models["decision_tree"]["breast"].fit(breast_trainX, breast_trainY)

models["decision_tree"]["pneumonia"] = DecisionTreeClassifier(criterion="gini", min_samples_split=3, min_samples_leaf=10)
models["decision_tree"]["pneumonia"].fit(pneumonia_trainX, pneumonia_trainY)
```

```python
print(f'breast train acc: {roc(models["decision_tree"]["breast"], breast_trainX, breast_trainY)}')
print(f'breast test acc: {roc(models["decision_tree"]["breast"], breast_testX, breast_testY)}')
print(f'breast val acc: {roc(models["decision_tree"]["breast"], breast_valX, breast_valY)}')
print()
print(f'pneumonia train acc: {roc(models["decision_tree"]["pneumonia"], pneumonia_trainX, pneumonia_trainY)}')
print(f'pneumonia test acc: {roc(models["decision_tree"]["pneumonia"], pneumonia_testX, pneumonia_testY)}')
print(f'pneumonia val acc: {roc(models["decision_tree"]["pneumonia"], pneumonia_valX, pneumonia_valY)}')
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

#### Hyperparameters to Optimize

- `penalty` - {`None`, `l2`, `l1`, `elasticnet`}
- `C` - Inverse of regularization strength
- `class_weight` - {`balanced`, `None`}

```python
params = itertools.product(
    (
        (None, "lbfgs"),
        ("l2", "lbfgs"),
        ("l1", "liblinear") 
    ),
    range(-6, 6),
    ("balanced", None)
)
params = list(params)

def eval_params(penalty, C, class_weight):
    penalty, solver = penalty
    model = LogisticRegression(penalty=penalty, solver=solver, C=10**C, class_weight=class_weight)
    return cross_val_score(model, breast_trainX, breast_trainY, scoring="roc_auc", cv=5, n_jobs=-1).mean()

results = Parallel(n_jobs=-1, verbose=10)(delayed(eval_params)(penalty, C, class_weight) for penalty, C, class_weight in params)
```

```python
best_score_index = results.index(max(results))
best_score = results[best_score_index]
best_params = params[best_score_index]

print(f"Params: {best_params}")
print(f"Score: {best_score}")
```

#### Model Accuracy

```python
models["logistic"]["breast"] = LogisticRegression(penalty="l1", solver="liblinear", C=10**5, class_weight=None)
models["logistic"]["breast"].fit(breast_trainX, breast_trainY)

models["logistic"]["pneumonia"] = LogisticRegression(penalty="l1", solver="liblinear", C=10**5, class_weight=None)
models["logistic"]["pneumonia"].fit(pneumonia_trainX, pneumonia_trainY)
```

```python
print(f'breast train acc: {roc(models["logistic"]["breast"], breast_trainX, breast_trainY)}')
print(f'breast test acc: {roc(models["logistic"]["breast"], breast_testX, breast_testY)}')
print(f'breast val acc: {roc(models["logistic"]["breast"], breast_valX, breast_valY)}')
print()
print(f'pneumonia train acc: {roc(models["logistic"]["pneumonia"], pneumonia_trainX, pneumonia_trainY)}')
print(f'pneumonia test acc: {roc(models["logistic"]["pneumonia"], pneumonia_testX, pneumonia_testY)}')
print(f'pneumonia val acc: {roc(models["logistic"]["pneumonia"], pneumonia_valX, pneumonia_valY)}')
```

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
```

#### Hyperparamters to Optimize

- `C` - regularization param
- `kernel` - {`poly`, `rbf`, `sigmoid`}
- `degree` (for `poly`) and `gamma` (for `rbf` and `poly` and `sigmoid`)

##### `Poly` Kernel

```python
# for poly
params = itertools.product(
    range(-6, 6),
    range(1, 8),
    ("scale", "auto")
)
params = list(params)

def eval_params(C, degree, gamma):
    model = SVC(C=10**C, kernel="poly", degree=degree, gamma=gamma, probability=True)
    return cross_val_score(model, breast_trainX, breast_trainY, scoring="roc_auc", cv=5, n_jobs=-1).mean()

results = Parallel(n_jobs=-1, verbose=10)(delayed(eval_params)(C, kernel, degree) for C, kernel, degree in params)
```

```python
best_score_index = results.index(max(results))
best_score = results[best_score_index]
best_params = params[best_score_index]

print(f"Params: {best_params}")
print(f"Score: {best_score}")
```

##### Other Kernels

```python
# for poly
params = itertools.product(
    range(-6, 6),
    ("linear", "rbf", "sigmoid"),
    ("scale", "auto")
)
params = list(params)

def eval_params(C, kernel, gamma):
    model = SVC(C=10**C, kernel=kernel, gamma=gamma, probability=True)
    return cross_val_score(model, breast_trainX, breast_trainY, scoring="roc_auc", cv=5, n_jobs=-1).mean()

results = Parallel(n_jobs=-1, verbose=10)(delayed(eval_params)(C, kernel, degree) for C, kernel, degree in params)
```

```python
best_score_index = results.index(max(results))
best_score = results[best_score_index]
best_params = params[best_score_index]

print(f"Params: {best_params}")
print(f"Score: {best_score}")
```

#### Model Accuracy

```python
models["svc"]["breast"] = SVC(C=10, kernel="rbf", gamma="scale", probability=True)
models["svc"]["breast"].fit(breast_trainX, breast_trainY)

models["svc"]["pneumonia"] = SVC(C=10, kernel="rbf", gamma="scale", probability=True)
models["svc"]["pneumonia"].fit(pneumonia_trainX, pneumonia_trainY)
```

```python
print(f'breast train acc: {roc(models["svc"]["breast"], breast_trainX, breast_trainY)}')
print(f'breast test acc: {roc(models["svc"]["breast"], breast_testX, breast_testY)}')
print(f'breast val acc: {roc(models["svc"]["breast"], breast_valX, breast_valY)}')
print()
print(f'pneumonia train acc: {roc(models["svc"]["pneumonia"], pneumonia_trainX, pneumonia_trainY)}')
print(f'pneumonia test acc: {roc(models["svc"]["pneumonia"], pneumonia_testX, pneumonia_testY)}')
print(f'pneumonia val acc: {roc(models["svc"]["pneumonia"], pneumonia_valX, pneumonia_valY)}')
```

## Analyze Model Performance

### Breast Dataset

```python
data = []

for model in models:
    for dataset_type in ("train", "test", "val"):
        data.append((
            model,
            dataset_type,
            roc(models[model]["breast"], eval(f"breast_{dataset_type}X"), eval(f"breast_{dataset_type}Y"))
        ))
        
data = pd.DataFrame(data, columns=["model", "type", "score"])
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="model", y="score", hue="type") \
    .set_title("ROC of Different Models for Breast")

# plt.savefig("roc_breast.png")
```
